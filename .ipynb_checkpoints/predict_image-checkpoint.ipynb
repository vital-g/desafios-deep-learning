# Fazendo a predição de uma unica foto

```python
# primeiro vamos improtar as bibliotecas necessarias
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from os import listdir
```

## defina o path da tua foto de amostra

```python
from os import listdir
dataset = 'path/to/your/sample'
dataset = 'images/tap.jpg'

dataset = 'dataset/train/tap/' +listdir('dataset/train/tap/')[0]
```

```python
# load and prepare the image
def load_image(filename):
    # load the image
    img = load_img(filename, target_size=(224, 224))
    # convert to array
    plt.imshow(img)
    img = img_to_array(img)
    # reshape into a single sample with 3 channels
    img = img.reshape(1, 224, 224, 3)
    # center pixel data
    img = img.astype('float32')
    return img

# load an image and predict the class

# load the image
img = load_image(dataset)
# load model
model_name = listdir('model/final_model/')[0]

model = load_model('model/final_model/' + model_name)

# predict the class
result = model.predict(img)

if result[0][0] > 0.5:
    print('tap')
else:
    print('klm')
```
